```
x = df.drop(['ID','ZIP Code','Personal Loan'],axis=1)
y = df['Personal Loan']
```

```
x.shape
```

```
y.shape
```

```
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=100)
```

```
x_train.shape
```

```
x_test.shape
```

```
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train_s = sc.fit_transform(x_train)
x_test_s = sc.fit_transform(x_test)
```

```
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train_s,y_train)
```

```
y_pred = lr.predict(x_test_s)
```

```
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```
confusion_matrix(y_test,y_pred)
```

```
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True,fmt='d');
```

```
accuracy_score(y_test,y_pred)
```

```
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(x_train_s,y_train)
```

```
y_pred = gnb.predict(x_test_s)
```

```
confusion_matrix(y_test,y_pred)
```

```
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True,fmt='d');
```

```
accuracy_score(y_test,y_pred)
```

```
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train_s,y_train)
```

```
y_pred = knn.predict(x_test_s)
```

```
confusion_matrix(y_test,y_pred)
```

```
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True,fmt='d');
```

```
accuracy_score(y_test,y_pred)
```

```
accuracy_rate = []
for i in range(1,40):
  knn = KNeighborsClassifier(n_neighbors=i)
  knn.fit(x_train_s,y_train)
  score = knn.score(x_test_s,y_test)
  accuracy_rate.append(score.mean())
```

```
plt.figure(figsize=(12,8))
plt.plot(range(1,40),accuracy_rate,marker='o',color='blue')
plt.xlabel('n_neighbors=i')
plt.ylabel('accuracy_rate')
plt.show()
```

```
from sklearn import tree
dtree = tree.DecisionTreeClassifier(criterion='entropy', max_depth=3)
dtree.fit(x_train_s,y_train)
```

```
y_pred = dtree.predict(x_test_s)
```

```
confusion_matrix(y_test,y_pred)
```

```
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True,fmt='d');
```

```
accuracy_score(y_test,y_pred)
```

```
dot_data = tree.export_graphviz(dtree)
from IPython.display import Image
import pydotplus
graph = pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
```

```
from sklearn.ensemble import RandomForestClassifier
```

```
from sklearn.model_selection import GridSearchCV
```

```
rf_cl = RandomForestClassifier(n_estimators=250, random_state=21)
```

```
from sklearn.model_selection import GridSearchCV
rf_cl = RandomForestClassifier(n_estimators=250, random_state=21)
grid_search = {'max_features':[4,5,6,7,8,9,10],'min_samples_split':[2,3,4,5,10],'n_estimators':range(0,250,50)}
grid_search = GridSearchCV(rf_cl,
                           param_grid = grid_search,
                           scoring = 'accuracy',
                           cv = 10,
                           n_jobs = -1)
```

```
grid_search.fit(x_train_s,y_train)
best_parameters = grid_search.best_params_
best_accuracy = grid_search.best_score_
```
